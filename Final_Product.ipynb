### *Does higher spending guarantee success in the MLB?*

![Infographic](Final_Infographic.png)

This infographic was made using Canva, and contains all three of the final visualizations I made in Notebook 2. It aims to answer the main question, "Can higher spending guarantee success in the MLB?". Moving down the infographic you can see we start from overall trend in the data, then we look at playoff teams, then world series winners. Moving chronologically like this made the most sense in my mind as we can look at three of the most important phases of the MLB season.
